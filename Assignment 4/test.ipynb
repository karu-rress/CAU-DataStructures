```python
from collections import OrderedDict
d = OrderedDict()

with open('randdict_utf8.txt', 'r') as f:
    for line in f:
        try:
            key, value = line.strip().split(' : ')
            d[key] = value
        except:
            d[key] = ''

d = OrderedDict(sorted(d.items(), key=lambda t: t[0]))
```

```python
d
```

```python
i, j = 0, 0

for key, value in d.items():
    if len(key) > i:
        i = len(key)
    if len(value) > j:
        j = len(value)
```
